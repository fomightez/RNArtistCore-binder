```python
# first we create a directory to store the scripts and output files
%mkdir binder_tests
```

### RNA 2D described from scratch in the script file

```python
s = '''rnartist {
  file = "binder_tests/demo1.svg"
  ss {
    rna {
      sequence = "CAACAUCAUACGUACUGCGCCCAAGCGUAACGCGAACACCACGAGUGGUGACUGGUGCUUG"
    }
    bracket_notation =
      "(((..(((..(((..(((((....)))))..)))..(((((....)))))..)))...)))"
  }
  theme {
    details_lvl = 5

    color {
      type="A"
      value = "#A0ECF5"
    }

    color {
      type="a"
      value = "black"
    }

    color {
      type="U"
      value = "#9157E5"
    }

    color {
      type="G"
      value = "darkgreen"
    }

    color {
      type="C"
      value = "#E557E5"
    }

  }
}
'''
```

```python
%store s >binder_tests/demo1.kts
```

```python
# we ask rnartistcore to draw the 2D as described in the script file
```

```python
%%bash
java -jar RNArtistCore/target/rnartistcore-0.2.7-SNAPSHOT-jar-with-dependencies.jar binder_tests/demo1.kts
```

```python
#we display the 2D saved in an SVG file
from IPython.core.display import SVG #<-- this line only needs to be run once per notebook
# the output file name is like: filename defined in the script + molecular chain name (default name is 'A'). 
SVG(filename='binder_tests/demo1_A.svg')
```

So that last result above doesn't show the nucleotide letters, presumably because of this statement in the current documentation under [The drawing algorithm element](https://github.com/fjossinet/RNArtistCore#the-drawing-algorithm-element):

>The size of the picture will fit the size of the drawing (with a minimum size of 1024x768 to see the residue characters).

However, in this case it seems barely missing that point because the following change in layout shows them:

```python
s = '''rnartist {
  file = "binder_tests/demo_lettersback.svg"
  ss {
      pdb {
        name = "B"
        file = "RNArtistCore/samples/1gid.pdb"
      }
  }
  theme {
    details_lvl = 5

    color {
      type="A"
      value = "#A0ECF5"
    }

    color {
      type="a"
      value = "black"
    }

    color {
      type="U"
      value = "#9157E5"
    }

    color {
      type="G"
      value = "darkgreen"
    }

    color {
      type="C"
      value = "#E557E5"
    }

  }
  layout {

        junction {
            type = 3
            out_ids ="nw ne"
        }

    }
}
'''
```

```python
%store s >binder_tests/demo_lettersback.kts
```

```python
%%bash
java -jar RNArtistCore/target/rnartistcore-0.2.7-SNAPSHOT-jar-with-dependencies.jar binder_tests/demo_lettersback.kts
```
